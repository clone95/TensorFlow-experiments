```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import os
import time
import sys
import math
import random
from sklearn.datasets import make_regression
from matplotlib import pyplot

# suppres Tensorflow GPU verbose output
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 
```

```python
# generate regression dataset
X, y = make_regression(n_samples=100, n_features=1, noise=50, random_state=5)
# plot regression dataset
pyplot.scatter(X,y)
pyplot.show()
```

```python
X_test, y_test = make_regression(n_samples=100, n_features=1, noise=50, random_state=6)
# plot regression dataset
pyplot.scatter(X_test,y_test)
pyplot.show()
```

```python
def build_model():
  model = keras.Sequential([
    layers.Dense(16, activation='relu', input_shape=[1]),
    layers.Dense(1)
  ])

  optimizer = tf.keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)

  model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mse'])
  return model

model = build_model()
```

```python
for i in range(1,10):
    model.fit(X, y, epochs=20, validation_split = 0.2, verbose=0)
    y_pred = model.predict(X_test)
    pyplot.scatter(X_test,y_pred)
    pyplot.show()
```
